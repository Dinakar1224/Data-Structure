```python
#1.Discuss string slicing and provide examples?
#Ans:String slicing in Python is a way to extract a portion of a string (a substring) using a specific syntax. The basic syntax for slicing is:
String[start:stop:step]
start: The starting index of the slice (inclusive). If omitted, it defaults to the beginning of the string (index 0).
stop: The ending index of the slice (exclusive). If omitted, it defaults to the end of the string.
step: The step value, which determines how much to increment the index. If omitted, it defaults to 1.
Examples of String Slicing
1.Basic Slicing
Ex-  Text = “Hello, World!”
   Print(text[0:5])  # Output: Hello
 Here, text[0:5] returns the substring from index 0 to index 4 (not including 5), which is ”Hello”.

2.Slicing with Default Start and Stop
Ex-Text = “Hello, World!”
   Print(text[:5])   # Output: Hello
   Print(text[7:])   # Output: World!
Here,text[:5]` extracts the substring from the start to index 4.
text[7:] extracts the substring from index 7 to the end.

3.Slicing with a Negative Index
Ex-Text = “Hello, World!”
   Print(text[-6:-1])  # Output: World
Negative indices count from the end of the string. text[-6:-1] extracts the substring starting from the 6th last character to the 2nd last character (not including the last one).

4.Slicing with a Step
Ex-Text = “Hello, World!”
   Print(text[0:12:2])  # Output: Hlo ol
  Here,text[0:12:2] extracts every second character from index 0 to index 11.

5.Reversing a String Using Slicing
Ex-Text = “Hello, World!”
   Print(text[::-1])  # Output: !dlroW ,olleH
   Here,Text[::-1] reverses the string by starting from the end and moving backward with a step of -1.

6.Omitting All Parameters
Ex-Text = “Hello, World!”
   Print(text[:])  # Output: Hello, World
Here,Text[:] returns the	 entire string, as no start, stop, or step values are specified.
```

```python
#2.Explain the key features of lists in Python
#Ans:Lists in Python are a versatile and widely used data structure.
#key features:
1.Ordered Collection: Lists maintain the order of elements as they are added. This means that the position of each element is preserved, and you can access elements by their index.
2.Indexing and Slicing: You can access elements using zero-based indexing, and extract sublists using slicing. For example, my_list[0] gives you the first element, and my_list[1:4] gives you a sublist from index 1 to 3.
3	Mutable: Lists are mutable, meaning you can change their content (modify, add, or remove elements) after the list has been created. For example, you can use my_list.append(5) to add an element to the end.
4.Dynamic Size: Lists can grow or shrink in size as needed. You don't need to specify their size in advance.
5.Can Contain Mixed Types: Lists can hold items of different data types, including integers, strings, and even other lists. For example, my_list = [1, 'pwskills', 3.14, [1, 2, 3]].
6.List Operations: You can perform various operations on lists, such as concatenation (list1 + list2), repetition (list * n), and membership testing (item in list).
7.Built-in Methods: Python lists come with a variety of useful methods, including:
-append(x): Adds an item x to the end of the list.
-extend(iterable): Extends the list by appending elements from an iterable.
-insert(i, x): Inserts an item x at a given position i.
-remove(x): Removes the first item from the list whose value is equal to x.
-pop([i]): Removes and returns the item at the given position i
in the list. If no index is specified, pop() removes and returns the last item.
-sort(): Sorts the items of the list in place.
-reverse(): Reverses the elements of the list in place.
-copy(): Returns a shallow copy of the list.
8.	List Comprehensions: Python supports list comprehensions, which provide a concise way to create lists based on existing lists. For example, [x**2 for x in range(10)] creates a list of squares for numbers 0 through 9.
9.	Nested Lists: Lists can contain other lists, which allows for the creation of complex data structures like matrices. For example, matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]].
```

```python
#3.Describe how to access, modify, and delete elements in a list with examples
#Ans: Accessing, modifying, and deleting elements in a Python list
1. Accessing Elements
You can access elements in a list using their index. Indexing in Python is zero-based, meaning the first element is at index 0.
Examples:
# Define a list
my_list = [10, 20, 30, 40, 50]

# Access elements by index
first_element = my_list[0]    # 10
second_element = my_list[1]   # 20
last_element = my_list[-1]    # 50 (negative index accesses elements from the end)

# Access a slice of the list
subset = my_list[1:4]         # [20, 30, 40] (from index 1 to 3, excluding index 4)

2. Modifying Elements
You can modify elements in a list by assigning a new value to a specific index. Lists being mutable allows this kind of operation.
Examples:
# Define a list
my_list = [10, 20, 30, 40, 50]

# Modify elements by index
my_list[1] = 25           # Change the second element from 20 to 25

# Modify a slice of the list
my_list[2:4] = [35, 45]   # Replace elements at index 2 and 3 with [35, 45]

# Check the updated list
print(my_list)            # Output: [10, 25, 35, 45, 50]
3. Deleting Elements
You can delete elements from a list using several methods: del, pop(), and remove().
Examples:
python
Copy code
# Define a list
my_list = [10, 20, 30, 40, 50]

# Delete an element by index using del
del my_list[2]            # Remove the element at index 2 (30)

# Delete an element by index using pop()
removed_element = my_list.pop(1)  # Remove and return the element at index 1 (20)
print(removed_element)          # Output: 20

# Delete the first occurrence of a value using remove()
my_list.remove(40)         # Remove the first occurrence of the value 40

# Check the updated list
print(my_list)            # Output: [10, 50]
```

```python
#4. Compare and contrast tuples and lists with examples.
#Ans:Tuples and lists are both fundamental data structures in Python used for storing collections of items. They have several similarities but also some key differences. Here’s a detailed comparison:
Similarities:
1.Ordered Collections: Both tuples and lists maintain the order of elements as they are added. You can access elements by index.
2.Indexing and Slicing: Both support indexing and slicing to access elements or sub-sections.
3.Can Contain Mixed Types: Both can store elements of different data types, including other tuples or lists.

Differences:
1.Mutability:
-Lists: Mutable. You can modify a list after its creation (add, remove, or change elements).
-Tuples: Immutable. Once a tuple is created, you cannot change its content. You can't add, remove, or modify elements.
Examples:
# Lists
my_list = [1, 2, 3]
my_list[0] = 10  # Modification is allowed
my_list.append(4)  # Adding elements is allowed

# Tuples
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This will raise a TypeError
# my_tuple.append(4)  # Tuples don't have append() method
1.Syntax:
-Lists: Defined with square brackets [].
-Tuples: Defined with parentheses ().
Examples:
# List
my_list = [1, 2, 3]
# Tuple
my_tuple = (1, 2, 3)

2.Performance:
-Lists: Generally slower than tuples due to their dynamic nature and support for more operations.
-Tuples: Generally faster and have a smaller memory footprint due to their immutability.

3.Use Cases:
-Lists: Suitable for collections of items that may need to be modified. Examples include collections of user input, dynamically changing data, or lists of results.
-Tuples: Suitable for fixed collections of items that shouldn’t be modified. Examples include coordinates, records, or function return values that shouldn’t be altered.

4.Methods:
-Lists: Have several built-in methods such as append(), extend(), insert(), remove(), pop(), sort(), and reverse().
-Tuples: Have fewer methods. Mainly count() and index(), which are used to count occurrences and find the index of elements.
Examples:
# List methods
my_list = [1, 2, 3]
my_list.append(4)     # [1, 2, 3, 4]
my_list.remove(2)     # [1, 3, 4]

# Tuple methods
my_tuple = (1, 2, 3, 1)
count_of_1 = my_tuple.count(1)  # 2
index_of_2 = my_tuple.index(2)  # 1
5.Use as Dictionary Keys:
-Lists: Cannot be used as dictionary keys because they are mutable and therefore unhashable.
-Tuples: Can be used as dictionary keys if they contain only immutable elements.
Examples:
# List as dictionary key (will raise an error)
# my_dict = {[1, 2]: 'value'}  # TypeError: unhashable type: 'list'

# Tuple as dictionary key
my_dict = {(1, 2): 'value'}  # Works fine

Summary
•	Lists are mutable, defined with [], and suited for collections that may change.
•	Tuples are immutable, defined with (), and suited for fixed collections that should not change.
```

```python
#5. Describe the key features of sets and provide examples of their use
#Ans:Key Features of Sets
1.Unordered: Sets do not maintain any order of elements. The items are stored in an unordered fashion, and there is no index associated with elements.
2.Unique Elements: Sets automatically remove duplicate items. Each element in a set must be unique.
3.Mutable: You can add or remove elements from a set after its creation. However, the elements themselves must be immutable (e.g., numbers, strings, tuples).
4.Dynamic Size: Sets can grow or shrink dynamically as elements are added or removed.
5.Set Operations: Sets support various mathematical set operations such as union, intersection, difference, and symmetric difference.
6.No Indexing or Slicing: Unlike lists or tuples, sets do not support indexing or slicing.
7.Efficient Membership Testing: Checking whether an element is in a set is generally faster compared to lists due to the underlying hash table implementation.
Examples of Using Sets
Creating Sets:
#Define a set
my_set = {1, 2, 3, 4, 5}

#Define an empty set
empty_set = set()

Adding and Removing Elements:
#Define a set
my_set = {1, 2, 3}

#Add an element
my_set.add(4)            #{1, 2, 3, 4}

#Remove an element
my_set.remove(2)         #{1, 3, 4}

#Remove an element if it exists (no error if the element is not present)
my_set.discard(5)        #{1, 3, 4} (no error)

#Pop an element (removes and returns an arbitrary element)
popped_element = my_set.pop()  #Could be 1, 3, or 4

#Clear all elements
my_set.clear()           #set()

Set Comprehensions:
# Define a set using set comprehension
squares = {x**2 for x in range(6)}  #{0, 1, 4, 9, 16, 25}

Using Sets with Dictionaries:
Sets can be useful for creating dictionary keys if they contain only immutable elements. For instance:
#Define a set of tuples as dictionary keys
my_dict = { (1, 2): 'a', (3, 4): 'b' }

#Access values using the set of tuples
print(my_dict[(1, 2)])  # Output: 'a'.
```

```python
#7. Describe how to add, modify, and delete items in a dictionary with examples
#Ans.1.Adding Items
You can add new key-value pairs to a dictionary by assigning a value to a new key.
Examples:
#Define a dictionary
my_dict = {'apple': 1, 'banana': 2}

#Add a new key-value pair
my_dict['orange'] = 3
print(my_dict)  # Output: {'apple': 1, 'banana': 2, 'orange': 3}

#2.Modifying Items
You can modify the value associated with an existing key by assigning a new value to that key.
Examples:
#Define a dictionary
my_dict = {'apple': 1, 'banana': 2}

#Modify the value for an existing key
my_dict['apple'] = 10

print(my_dict)  # Output: {'apple': 10, 'banana': 2}

#3.Deleting Items
You can delete items from a dictionary using several methods: del, pop(), and popitem().
Examples:
1.Using del:
-Removes a key-value pair by specifying the key.
# Define a dictionary
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}

# Delete an item by key
del my_dict['banana']
print(my_dict)  # Output: {'apple': 1, 'orange': 3}

2.Using pop():
-Removes an item by key and returns the value. If the key is not found, it raises a KeyError unless a default value is provided.
# Define a dictionary
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}

# Remove an item by key and get the value
value = my_dict.pop('orange')

print(value)    # Output: 3
print(my_dict)  # Output: {'apple': 1, 'banana': 2}

#Remove an item by key with a default value if the key is not found
value = my_dict.pop('grape', 'Not Found')
print(value)    # Output: Not Found

3.Using popitem():
-Removes and returns the last inserted key-value pair as a tuple. This method does not require a key and raises a KeyError if the dictionary is empty.
#Define a dictionary
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}

#Remove the last inserted item
item = my_dict.popitem()

print(item)     # Output: ('orange', 3)
print(my_dict)  # Output: {'apple': 1, 'banana': 2}

4.Using clear():
-Removes all items from the dictionary, leaving it empty.
#Define a dictionary
my_dict = {'apple': 1, 'banana': 2}

#Remove all items
my_dict.clear()
print(my_dict)  # Output: {}
```
